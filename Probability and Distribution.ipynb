```python
#possion distrubution using numpy
q_size = np.random.poisson(3.6, 100) #with the mean of 3.6 and probability of getting 100 number in a queue
```

```python
#lets calculate the rod of 100m with 2 standard deviation and we took a sample of 100 rod
#its as normal distribution, gaussian, continue data 
```

```python
possible_size = np.random.normal(100, 2, 10000)
```

```python
#for normal distribution we use histogram to show
sns.set_theme(style='whitegrid')
sns.histplot(x=possible_size, kde=True) #kernal density line
plt.show()
```

```python
#here (+1 sigma, -1 sigma )cover almost 68% data
sum((possible_size > 98)&(possible_size < 102))
```
